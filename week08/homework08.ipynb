```python
start_date = []
end_date = []
phone = []

for _, i in df["notes"].iteritems():
    start = i.split("end_date")
    start_date = start[0].replace("start ","")
    start.append(start_date)
    pattern = re.compile('P#|pn|p#|p/n')
    
    if (pattern.search(start[1])):
        mo = pattern.search(start[1])
        print(mo.group())
        
        end = start[1].split(mo.group())
        print(end)
        end_date.append(end[0])
        phone.append(end[1])
    else:
        end_date.append(start[1])
        phone.append(None)
    
```

```python
df[["startdate", "enddate"]] = df["notes"].str.split(r"en\s?dt|end\s?dt|end\s?date|en\s?date|ed\s?date|ed\s?dt|EN\s?DT|END\s?DT|END\s?DATE|ED\s?DATE|ED\s?DATE|ED\s?DT|end\s?Dt", expand=True)
df[["phone_number", "startdate"]] = df["startdate"].str.split(r"st\s?dt|start\s?dt|start\s?date|st\s?date|ST\s?DT|START\s?DT|START\s?DATE|ST\s?DATE|st\s?Dt", expand=True)
df[["enddate", "phone_number"]] = df["enddate"].str.split(r"pn|p#|p/n|P#", expand=True)
df[["phone_number", "other"]] = df["phone_number"].str.split(r"x", expand=True)
```

```python
df['startdate']=df['startdate'].replace(':','',regex=True)
df['startdate']=df['startdate'].replace(';','',regex=True)
df['startdate']=df['startdate'].replace(r'-$','',regex=True)
df['startdate']=df['startdate'].replace(r'/$','',regex=True)

df["startdate"] = df["startdate"].str.lower()
df['startdate']=df['startdate'].replace(r'/','-',regex=True)

df['enddate']=df['enddate'].replace(':','',regex=True)
df['enddate']=df['enddate'].replace(';','',regex=True)
df['enddate']=df['enddate'].replace(r'-$','',regex=True)
df['enddate']=df['enddate'].replace(r'/$','',regex=True)

df["enddate"] = df["enddate"].str.lower()
df['enddate']=df['enddate'].replace(r'/','-',regex=True)

df.startdate = df.startdate.str.replace(' ', '')
df.enddate = df.enddate.str.replace(' ', '')
```

# Create start date and end date columns

```python
pattern = re.compile(r'(\d{1,2}(?:-|\/)?\d{4})')
for x in range(len(df['startdate'])):
    if pattern.match(df['startdate'][x]):
        df['startdate'][x] = '01-' + df['startdate'][x]
        
df['startdate'] = df['startdate'].str.split('-').str[::-1].str.join(' ')

pattern = re.compile(r'(\d{2}) (\d{2})$')


for x in range(len(df['startdate'])):
    if pattern.match(df['startdate'][x]):

        df['startdate'][x] = '20' + df['startdate'][x]
        
df['startdate'] = df['startdate'].str.split(' ').str.join('-')
df['startdate'] = pd.to_datetime(df.startdate, dayfirst=True, errors='coerce')
```

```python
pattern = re.compile(r'(\d{1,2}(?:-|\/)?\d{4})')

for x in range(len(df['enddate'])):
    
    if pattern.match(df['enddate'][x]):
        
        df['enddate'][x] = '01-' + df['enddate'][x]
       
        
df['enddate'] = df['enddate'].str.split('-').str[::-1].str.join(' ')


pattern = re.compile(r'(\d{2}) (\d{2})$')


for x in range(len(df['enddate'])):
    if pattern.match(df['enddate'][x]):
        
        df['enddate'][x] = '20' + df['enddate'][x]
        
        
df['enddate'] = df['enddate'].str.split(' ').str.join('-')

df['enddate'] = pd.to_datetime(df.enddate, dayfirst=True, errors='coerce')
```
