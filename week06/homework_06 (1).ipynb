```python
file_url = 'https://github.com/msaricaumbc/DS_data/blob/master/bigfile.csv.zip?raw=true'
file_name = 'bigfile.zip'
```

```python
# function to download dataset
def download_file(file_name, url):
    res = request.urlopen(url)
    with open(file_name,'wb') as file:
        file.write(res.read())
```

```python
# function to unzip the downloaded zip file
def unzip(file_name, path='./'):
    # opening the zip file in READ mode 
    with ZipFile(file_name, 'r') as zip: 
        # printing all the contents of the zip file 
        zip.printdir() 

        # extracting all the files 
        print('Extracting all the files now...') 
        zip.extractall(path = path) 
        print('Done!')
```

```python
#reading line by line from CSV file
#appending the ages of the people to another list
#appending all the occupations to another list
import csv

age=[]
occupation=[]

with open('bigfile.csv', 'r') as file:
    while(True):
        line = file.readline()
        if not line:
            break
        line2 = line.split(',')
        age.append(int(line2[3]))
        occupation.append(line2[4])
file.close()
```
