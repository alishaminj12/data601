```python
from urllib import request
```

```python
def download_file(file_name, url):
    res = request.urlopen(url)
    with open(file_name,'wb') as file:
        file.write(res.read())
```

```python
file_url = 'https://github.com/msaricaumbc/DS_data/blob/master/bigfile.csv.zip?raw=true'
file_name = 'bigfile.zip'
```

```python
#reading line by line from CSV file
#appending the ages of the people to another list
#appending all the occupations to another list
import csv
age=[]
occupation=[]
with open('bigfile.csv', 'r') as file:
    while(True):
        line = file.readline()
        if not line:
            break
        line2 = line.split(',')
        age.append(int(line2[3]))
        occupation.append(line2[4])
file.close()
```

```python
#2.Create a histogram of ages

from matplotlib import pyplot as plt
plt.figure(figsize=(10,8))
plt.title("Histogram Plot For Ages")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.hist(age)
plt.show()
```
