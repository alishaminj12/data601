#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(data = diamonds, x="carat", y="price", hue='clarity')
```

```python
Prices are increased as the carat increases between theone and two
```

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True)
plot.ax.tick_params(labelrotation = 90)
```
