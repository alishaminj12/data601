```python
multiple_grouped = flights \
   .groupby(['day', 'month']) \
   .agg(
    flight_min= ('time_hour','min'), 
    flight_max= ('time_hour','max')
    )

multiple_grouped
   
```

```python
def my_delay_func(group):
    return group.dep_delay.mean() > 60

## use my_delay_func to .filter the original dataframe
filtered_by_delay = flights.groupby('dest').filter(my_delay_func)
filtered_by_delay
```

```python
# return all the flights with more than 60ins delay
flights[flights.dep_delay > 60]
```

```python
agg_dict = {
    'dest': 'count',
    'dep_delay': 'mean' 
}
```

```python
## use the dictionary we created above in .aggregate method
aggregated_df = flights.groupby('dest').aggregate(agg_dict)
aggregated_df
```
