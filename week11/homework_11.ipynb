## Scrape data

```python
page_content = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)#IMF_projections_for_2020_through_2027")
page_content
```

```python
type(page_content)
```

```python
country_imf = page_content[5]
```

```python
country_imf
```

```python
country_imf.rename(columns={'Country (or dependent territory)': 'Country'}, inplace=True)
```

```python
country_imf.replace({'NaN': None})
```

## Append to the dataframe

```python
mergefile = df1[['Country', 'Capital', 'Area','Population']].copy()
```

```python
mergefile.head()
```

```python
df2 = mergefile.merge(country_imf,on='Country')
```

```python
df2
```

```python
df2.drop(['2020','2021'], axis = 1)
```

## Calculate average

```python
# Calculating Percentage
df2['percent'] = (df2['2027'] - df2['2022'] / 
                  df2['2022'].sum()) * 100
```

```python
df2
```

## Sort in descending order.

```python
df2.sort_values(by='percent', ascending=False)
```
